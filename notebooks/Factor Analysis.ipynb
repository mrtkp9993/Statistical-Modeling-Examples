```python
import pandas as pd
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
import  rpy2.robjects.packages as rpackages
pandas2ri.activate()
psych = rpackages.importr("psych")
%load_ext rpy2.ipython
```

```python
# http://staff.bath.ac.uk/pssiw/stats2/page16/page16.html
df = pd.read_csv("data/bfi.csv", index_col=0)
df = df.iloc[:, 0:25]
df.head()
```

```python
# pandas df to r df
r_df = pandas2ri.py2ri(df)
```

```python
# perform paralel analysis
psych.fa_parallel(r_df)
```

```python
# factor analysis
fa_out = psych.fa(r_df, nfactors=6, fm="pa", maxiter=300)
```

```python
# show diagram
psych.fa_diagram(fa_out)
```
